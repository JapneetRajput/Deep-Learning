```python
import matplotlib.pyplot as plt
import numpy as np
x=[0,0,1,1]
y=[0,1,0,1]
```

AND gate

```python
ans_and=[]
_and=[]
w1=1
w2=1
theta=2
for i in range(len(x)):
  _and.append(x[i]*w1+y[i]*w2)
  if(x[i]*w1+y[i]*w2>=theta):
    ans_and.append(1)
  else:
    ans_and.append(0)

print(ans_and)
```

```python
firing_points = []
for num in _and:
    indices = np.where(_and == num)[0]
    if len(indices) > 0:
        i = indices[0]
        point = (x[i], y[i])
        firing_points.append(point)
```

```python
x1_values = [point[0] for point in firing_points]
x2_values = [point[1] for point in firing_points]
x1 = np.linspace(0, 1)
y1 = -1*x1 + 1.5
plt.plot(x1, y1, linestyle='dashed', color='black')
plt.scatter(x, y, marker="x")
plt.scatter([0,0,1],[0,1,0], color='blue',marker='x')
plt.scatter(x1_values, x2_values, color='red')
```
